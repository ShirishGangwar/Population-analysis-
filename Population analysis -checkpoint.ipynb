```python
t=df['Urban Pop %']
t2=list(t)
t2
```

```python
k=[]
for i in t2:
    r=""
    for j in i:
        if(j.isdigit()):
            r=r+j
    k.append(r)
k
    
```

```python
df['Urban Pop %']=k
```

```python
df['Urban Pop %']=df['Urban Pop %'].replace('',0)
```
